```python
import numpy as np
import matplotlib.pyplot as plt
```

##Sin Function

```python
x=np.arange(-np.pi,np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

##Cos Function

```python
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

#Unit Impulse *Function*

```python
x=np.arange(-10,11,1)
impulse=[]
for i in x:
    if i==0:
        impulse.append(1)
    else:
        impulse.append(0)
plt.stem(x,impulse)
plt.show()
```

## Unit Step Function

```python
x=np.arange(-10,11,1)
step=[]
for i in x:
    if i>=0:
        step.append(1)
    else:
        step.append(0)
plt.stem(x,step)
plt.show()
```

## Unit Ramp Function

```python
x=np.arange(-10,11,1)
ramp=[]
for i in x:
    if i>=0:
        ramp.append(i)
    else:
        ramp.append(0)
plt.stem(x,ramp)
plt.show()
```

## Step + Impulse

```python
stepimpulse=[]
for i in range(len(step)):
    stepimpulse.append(step[i]+impulse[i])
plt.stem(x,stepimpulse)
plt.show()
```
