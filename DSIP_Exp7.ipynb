```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('Sunflower.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

mask = (gray_image >= 50) & (gray_image <= 200)
masked_image = np.zeros_like(gray_image)
masked_image[mask] = gray_image[mask]

min_val = 50
max_val = 200
stretched_image = np.clip((masked_image - min_val) * (255 / (max_val - min_val)), 0, 255).astype(np.uint8)

plt.figure(figsize=(12, 12))

plt.subplot(4, 2, 1)
plt.title('Original Image (Color)')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(4, 2, 2)
plt.title('Original Image Histogram')
plt.hist(gray_image.ravel(), bins=256, range=(0, 256))
plt.xlim([0, 256])

plt.subplot(4, 2, 3)
plt.title('Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(4, 2, 4)
plt.title('Grayscale Image Histogram')
plt.hist(gray_image.ravel(), bins=256, range=(0, 256))
plt.xlim([0, 256])

plt.subplot(4, 2, 5)
plt.title('Masked Image (50-200)')
plt.imshow(masked_image, cmap='gray')
plt.axis('off')

plt.subplot(4, 2, 6)
plt.title('Masked Image Histogram')
plt.hist(masked_image.ravel(), bins=256, range=(0, 256))
plt.xlim([0, 256])

plt.subplot(4, 2, 7)
plt.title('Stretched Image')
plt.imshow(stretched_image, cmap='gray')
plt.axis('off')

plt.subplot(4, 2, 8)
plt.title('Stretched Image Histogram')
plt.hist(stretched_image.ravel(), bins=256, range=(0, 256))
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
