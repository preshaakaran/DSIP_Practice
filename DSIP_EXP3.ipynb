```python
import numpy as np

n = int(input("Enter the length of the signal a"))

print("Enter the values of signal a")
circular_a = []
for i in range(n):
  circular_a.append(int(input()))

n = int(input("Enter the length of the signal b"))
circular_b = []
for i in range(n):
  circular_b.append(int(input()))

circular_convolve = np.fft.ifft(np.fft.fft(circular_a) * np.fft.fft(circular_b))

print(circular_convolve)
```

```python

plt.subplot(3, 1, 1)
plt.stem(circular_a, use_line_collection=True)
plt.title('Signal a')

plt.subplot(3, 1, 2)
plt.stem(circular_b, use_line_collection=True)
plt.title('Signal b')

plt.subplot(3, 1, 3)
plt.stem(circular_convolve, use_line_collection=True)
plt.title('Convolved Signal')

plt.tight_layout()
plt.show()
```
