```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("bee.png")

# 2. Convert Color image to Gray Scale Image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Find the highest range(L-1) of gray level in the range [0, L-1] from the image
L = 256  # Assuming 8-bit grayscale

# 4. Perform Image Negative
negative_img = 255 - gray_img

# 5. Take Threshold values a and b as input
a = int(input("Enter threshold value a: "))
b = int(input("Enter threshold value b: "))

# 6. Apply gray level slicing with and without preserving the background
def gray_level_slicing(img, a, b, preserve_background=True):
    output = np.zeros_like(img)
    if preserve_background:
        output[np.logical_and(a <= img, img <= b)] = 255
    else:
        # Use np.where to get the indices of True values in the boolean array
        row_indices, col_indices = np.where(np.logical_and(a <= img, img <= b))
        # Assign values from img to output using the indices
        output[row_indices, col_indices] = img[row_indices, col_indices]
    return output

sliced_img_with_background = gray_level_slicing(gray_img, a, b, True)
sliced_img_without_background = gray_level_slicing(gray_img, a, b, False)

# 7. Take Threshold value T as input.
T = int(input("Enter threshold value T: "))

# 8. Apply thresholding on input image
_, thresh_img = cv2.threshold(gray_img, T, 255, cv2.THRESH_BINARY)

# 9. Conclude by specifying the applications where these operations can be used
print("Applications of these operations:")
print("- Image enhancement: Negative, gray level slicing, and thresholding can be used to enhance image contrast or extract specific features.")
print("- Image segmentation: Thresholding is commonly used to separate objects from the background in images.")
print("- Image analysis: These operations can be used for various image analysis tasks like object detection, feature extraction, and pattern recognition.")

# Display the images using cv2_imshow
print("Original Image")
cv2_imshow(img)
print("Gray Scale Image")
cv2_imshow(gray_img)
print("Negative Image")
cv2_imshow(negative_img)
print("Sliced Image with Background")
cv2_imshow(sliced_img_with_background)
print("Sliced Image without Background")
cv2_imshow(sliced_img_without_background)
print("Thresholded Image")
cv2_imshow(thresh_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
