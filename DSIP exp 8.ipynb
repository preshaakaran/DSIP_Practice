```python
import numpy as np
```

```python
import cv2
img = cv2.imread('image.jpeg')
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
```

```python
total_pixels = gray_img.size
prob = hist / total_pixels
```

```python
cdf = prob.cumsum()
```

```python
cdf_normalized = (cdf * 255).astype('uint8')
```

```python
cdf_rounded = np.round(cdf_normalized)
```

```python
import matplotlib.pyplot as plt

# Equalized image
equalized_img = cdf_rounded[gray_img]

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Input Gray Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(gray_img.ravel(), bins=256, range=[0, 256], color='black')
plt.title('Histogram of Input Image')

plt.subplot(2, 2, 3)
plt.imshow(equalized_img, cmap='gray')
plt.title('Output Equalized Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(equalized_img.ravel(), bins=256, range=[0, 256], color='black')
plt.title('Histogram of Output Image')

plt.tight_layout()
plt.show()
```
