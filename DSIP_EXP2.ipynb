```python
import numpy as np
import matplotlib.pyplot as plt

f = 9000

t = np.arange(0, 2e-4, 7e-6)

b = np.cos(2*np.pi*f*t)

plt.figure(1)
plt.plot(t, b, "g")
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Continuous Signal')
plt.show()
```

```python
ts = np.arange(0, 2e-4, 7e-6)

c = np.cos(2*np.pi*f*ts)
plt.figure(2)
plt.stem(ts, c, "k")
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Discrete Signal')
plt.show()
```
