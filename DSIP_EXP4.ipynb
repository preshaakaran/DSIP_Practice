```python
import numpy as np
import matplotlib.pyplot as plt

def get_signal(prompt):
    return np.array(list(map(float, input(prompt).split())))

def cross_correlation(x, y):
    return np.correlate(x, y, mode='full')

def auto_correlation(x):
    return np.correlate(x, x, mode='full')

signal1 = get_signal("Enter the first signal (space-separated values): ")
signal2 = get_signal("Enter the second signal (space-separated values): ")

cross_corr = cross_correlation(signal1, signal2)

auto_corr = auto_correlation(signal1)

plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.stem(signal1, use_line_collection=True)
plt.title('Signal 1')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.stem(signal2, use_line_collection=True)
plt.title('Signal 2')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 3)
plt.stem(cross_corr, use_line_collection=True)
plt.title('Cross-Correlation of Signal 1 and Signal 2')
plt.xlabel('Lag')
plt.ylabel('Correlation')

plt.subplot(4, 1, 4)
plt.stem(auto_corr, use_line_collection=True)
plt.title('Auto-Correlation of Signal 1')
plt.xlabel('Lag')
plt.ylabel('Correlation')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_signal(prompt):
    return np.array(list(map(float, input(prompt).split())))

signal1 = get_signal("Enter the first signal (space-separated values): ")
signal2 = get_signal("Enter the second signal (space-separated values): ")

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(signal1, use_line_collection=True)
plt.title('Signal 1')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.xlim(0, 9)

plt.subplot(3, 1, 2)
plt.stem(signal2, use_line_collection=True)
plt.title('Signal 2')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.xlim(0, 9)

convolved_signal = np.convolve(signal1, signal2)

plt.subplot(3, 1, 3)
plt.stem(convolved_signal, use_line_collection=True)
plt.title('Convolved Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.xlim(0, 9)

plt.tight_layout()
plt.show()
```
